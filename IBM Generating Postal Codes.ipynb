```python
tdf=df.groupby('Postal Code')['Neighbourhood'].apply(lambda x: "%s" % ', '.join(x))
tdf=tdf.reset_index(drop=False)
tdf.rename(columns={'Neighbourhood':'NNeighboUrhood'},inplace=True)
ndf = pd.merge(df, tdf, on='Postal Code')
ndf.drop(['Neighbourhood'],axis=1,inplace=True)
ndf.drop_duplicates(inplace=True)
ndf.rename(columns={'NNeighboUrhood':'Neighborhood'},inplace=True)
ndf.rename(columns={'Postal Code':'PostalCode'},inplace=True)
ndf
```
